```python
from transformers import pipeline
```

```python
#This code calls pipeline function, specifies its type, and defines the pre-trained as a keyword argument.
# In our case we are using GPT-Neo with 2.7 billion parameters. You may want to decrease the size depending on the speed of your computer.

gen = pipeline('text-generation', model ='EleutherAI/gpt-neo-2.7B')
```

```python
# to generate a text based on context, you will need to specify that context so that model can understand and generate text according to that context.
# Create a variable named context and write a sentence in string format.

context = "Deep Learning is a sub-field of Artificial Intelligence."

#First argument of the output variable is the context we specified as a string format. Second keyword argument, max_length defines the word count of the output.


output = gen(context, max_length=100, do_sample=True, temperature=0.9)
```

```python
#store the output data in a file

with open('dl.txt', 'w') as f:
       f.write(str(output))
```
